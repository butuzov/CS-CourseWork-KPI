## 7.1

\begin{bmatrix}5&2\\8&3\\1&4\\3&6\\2&5\end{bmatrix}

Для знаходження стратегії гравця 2 ми обираємо верхню оминаючу криву і з неї беремо дві пари що надають мінімальну точку.


$$
\begin{bmatrix}
8 & 3 \\ 
3 & 6  
\end{bmatrix} \Rightarrow 
\begin{array}{r|l}
8y_1+3y_2 & = u_2 \\
3y_1+6y_2 & = u_2 \\
y_1 + y_2 & = 1 \\
\end{array} \Rightarrow 
\begin{array}{r|l}
5y_1+-3y_2 & = 0 \\ 
y_1 + y_2 & = 1 \\
\end{array} \Rightarrow  
y_1 = \dfrac{3}{8} \quad y_2=\dfrac{5}{8} \quad u_2 = 4\dfrac{7}{8}
$$

Для знаходження стратегії гравця 1 ми транспонууємо матрицю і повторюємо усі операції.


$$
\begin{bmatrix}
8 & 3 \\ 
3 & 6  
\end{bmatrix} \Rightarrow 
\begin{array}{r|l}
8x_1+3x_2 & = u_1 \\
3x_1+6x_2 & = u_1 \\
x_1 + x_2 & = 1 \\
\end{array} \Rightarrow 
\begin{array}{r|l}
5x_1+-3x_2 & = 0 \\ 
x_1 + x_2 & = 1 \\
\end{array} \Rightarrow  
x_1 = \dfrac{3}{8} \quad x_2=\dfrac{5}{8} \quad u_1 = 4\dfrac{7}{8}
$$

```python
m=np.array([[8, 3], [3, 6]])
m
```

```python
np.multiply(m, [3/8,5/8]).sum(axis=1).reshape((-1, 1))
```

```python
np.multiply(m.T, [3/8,5/8]).sum(axis=1).reshape((-1, 1))
```

## 7.2

$$\begin{bmatrix}2, 3, 1, 6\\ 5,4,3,2\end{bmatrix} $$


Для знаходження стратегії гравця 1 ми обираємо нижню оминаючу криву і з неї беремо дві пари що надають максимальну точку.

$$
\begin{bmatrix}
1 & 3 \\ 
6 & 2  
\end{bmatrix} \Rightarrow
\begin{array}{r|l}
 x_1+3x_2 & = u_1\\
6x_1+2x_2 & = u_1 \\
x_1 + x_2 & = 1 \\
\end{array} \Rightarrow  
\begin{array}{r|l}
-5x_1+x_2 & = 0 \\ 
x_1 + x_2 & = 1 \\
\end{array} \Rightarrow  
x_1 = \dfrac{1}{6} \quad  x_2=\dfrac{5}{6} \quad  u_1 = 2\dfrac{2}{3}
$$

Для знаходження стратегії гравця 2 ми транспонууємо матрицю  і повторюємо усі операції.


$$
\begin{bmatrix}
1 & 6 \\ 
3 & 2  
\end{bmatrix} \Rightarrow 
\begin{array}{r|l}
 y_1+6y_2 & = u_2 \\
3y_1+2y_2 & = u_2 \\
y_1 + y_2 & = 1 \\
\end{array} \Rightarrow 
\begin{array}{r|l}
-2y_1+4y_2 & = 0 \\ 
y_1 + y_2 & = 1 \\
\end{array} \Rightarrow  
y_1 = \dfrac{4}{6} \quad y_2=\dfrac{2}{6} \quad u_2 = 2\dfrac{2}{3}
$$

```python
m = np.array([[1, 3], [6, 2]])
m
```

```python
np.multiply(m, [1/6,5/6]).sum(axis=1).reshape((-1, 1))
```

```python
np.multiply(m.T, [4/6,2/6]).sum(axis=1).reshape((-1, 1))
```
