  1. при відсутності коаліції знайти найкращі стратегії обох гравців і відповідні виграші.
  1. у кооперативній грі двох осіб вияснити доцільність створення коаліції і знайти точку Неша.
  1. перевірити вплив дії погроз, визначити арбітражний розв’язок у випадку дії погроз і порівняти із точкою Неша.

### Гарантовані виграші обох гравців за відсутності кооаліції

$$ u^* = max X^TAY \qquad v^* = maxX^TBY $$

**Гравець A**

$$
\begin{bmatrix}
 4 & 1 \\ 
-3 & 8 
\end{bmatrix}^T \Rightarrow 
\begin{bmatrix}
 4 & -3 \\ 
 1 & 8
\end{bmatrix} \Rightarrow 
\begin{array}{r|l}
4x_1-3x_2 & = u^* \\
x_1+8x_2 & = u^* \\
x_1 + x_2 & = 1 \\
\end{array} \Rightarrow 
\begin{array}{r|l}
3x_1-11x_2 & = 0 \\ 
x_1 + x_2 & = 1 \\
\end{array}
$$ 

$$x_1 = \dfrac{11}{14} \quad x_2=\dfrac{3}{14} \quad u^* = \dfrac{5}{2}$$

**Гравець B**
$$
\begin{bmatrix}
10 & -4 \\ 
-1 & 4 
\end{bmatrix} \Rightarrow 
\begin{array}{r|l}
6y_1-4y_2 & = v^* \\
-1y_1+4y_2 & = v^* \\
y_1 + y_2 & = 1 \\
\end{array} \Rightarrow 
\begin{array}{r|l}
5y_1 & = 0 \\ 
y_1 + y_2 & = 1 \\
\end{array} 
$$ 

$$y_1 = 0 \quad y_2=1 \quad v^* = 4$$

### Доцільність створення коаліції. ТочкаНеша.

Вважаємо що маємо гру з лінійною-трансферабельною користніст. Оскільки для двох стратегій рівноваги $x_1=[1;0], y_1=[1;0]$ та $x_2=[0;1], y_2=[0;1]$ суми виграшів не однакові $ (4+6) != (8+4)$, то перетворєюємо маштаб по осі $v: v'=v\alpha$ так щоб корисності гравців ділились і передавались відношенням `1:1`:

$$
A[1;0]- B[1;0]\alpha = A[0;1]- B[0;1]\alpha \quad \Rightarrow \quad 
    4 - 6\alpha = 8 + 4\alpha \quad \Rightarrow \quad 
    \alpha = 2
$$


Знайдемо "перетворену" матрицю

$$
B' = B * \alpha = \begin{bmatrix}
 12 & 8 \\
 2 &  8 \\
\end{bmatrix}
$$


<div class="alert alert-danger">Данний графік - не привильний, точка F має міститись в координаті (-3;2)</div>

$R$ - максимальна можлива корисність яку ці два гравці можуть отримати в результаті кооперативної гри.

$$ R = (u'+v') = (A[0;0]+B'[0;0]) = (A[1;1]+B'[1;1]) = (4 + 12) = 16$$

$$v^{*'} =   v^{*}  \times \alpha = 4 \times 2 = 8$$

$$\overline{u} = 
    \dfrac{(u^* - v^{*'})+R}{2} = 
    \dfrac{( \dfrac{2}{5} - 8 )+16}{2} =
    \dfrac{21}{4}  \qquad \overline{v}' = 
    \dfrac{(v^{*'} - u^*)+R}{2} = 
    \dfrac{(8 - \dfrac{2}{5})+16}{2} =
    \dfrac{43}{4} $$

$$\overline{v} = \dfrac{\overline{v}'}{\alpha} =  \dfrac{\dfrac{43}{4}}{2} = \dfrac{43}{8}$$

### Гра в умовах погроз

Будуємо матрицю $A-B'$ розвязок цієї хадачі для гравця 1.

$$
\begin{bmatrix}
 -8 & -7 \\
 -5 & 0
\end{bmatrix}^T \Rightarrow 
\begin{bmatrix}
 8 & -5 \\
 7 & 0  \\
\end{bmatrix} \Rightarrow 
\begin{array}{r|l}
8x_1-5x_2 & = X_y^T(A-B)Y_y \\
7x_1+0x_2 & = X_y^T(A-B)Y_y \\
x_1 + x_2 & = 1 \\
\end{array} \Rightarrow 
\begin{array}{r|l}
-1x_1-5x_2 & = 0 \\ 
x_1 + x_2 & = 1 \\
\end{array}
$$ 

$$x_1 = \dfrac{5}{5} \quad x_2=-\dfrac{1}{4} \quad X_y^T(A-B)Y_y = -\dfrac{35}{4}$$


$$
\overline{u}_y = \dfrac{ X_y^T(A-B)Y_y + r}{2} = \dfrac{-\dfrac{35}{4}+16}{2}  = \dfrac{29}{8} \qquad 
\overline{v}_y' = \dfrac{-X_y^T(A-B)Y_y + r}{2} = \dfrac{\dfrac{35}{4}+16}{2} = \dfrac{99}{8}$$

$$\overline{v}_y = \dfrac{\overline{v}'_y}{\alpha} =  \dfrac{99}{8} / 2 = \dfrac{99}{16}$$
